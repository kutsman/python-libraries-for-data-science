### 1. Для чего и в каких случаях полезны различные варианты усреднения для метрик качества классификации: micro, macro, weighted?

$Micro$ усреднение - доля верных срабатываний для всех классов, и соответствует метрике Accuracy

$\frac{1}{\large l}\sum_{i=1}^{\large l} [a(x_{i}) = y_{i}]$

$micro = \frac{TP+TN}{TN+TP+FN+FP}$

$Marco$ усреднение - это средне арифметическое метрик всех классов, т.е. сначала ищем метрику по каждому классу отдельно и далее берем от них среднее арифметическое.

$macro = \frac{1}{L}\sum_{i=0}^{L} m_{i}$, <br><br>$L $ - количество классов <br>$m_{i}$ - метрика для $i$ класса

$Weighted$ ищется как взвещенное среднее арифметическое, т.е. находим метрику и умножаем на вес (количество объектов с целевым классом / общее количество объектов)

$weighted = \sum_{i=0}^{L} \frac{n_{i}}{N} m_{i}$, <br><br>$\large l $ - количество классов <br>$m_{i}$ - метрика для $i$ класса<br>$n_{i}  $ - количество объектов $i$ класса<br>$N$ - количество объектов в выборке

Разные усреднения применяются когда у нас много классов

а) Micro используется когда идеальный баланс классов (т.е. когда нет дисбаланса классов)
<br>b) Macro используется когда нужно дать всем классам одинаковый вес (и есть дисбаланс классов), в независимости от того, сколько их в DF.
<br>с) Weighted используется когда нужно дать классам вес согласно их количествую и есть дисбаланс классов.

### 2. В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?

Модели отличаются друг от друга по нескольким критериям:

<br>1) симметричность деревьев, 
<br>2) метод разбиения объектов, 
<br>3) обработка категориальных признаков, 
<br>4) интерпретация пропущенных значений, 
<br>5) и обработка текстовых признаков. 

В CatBoost деревья симметричные на каждом уровне, в LightGBM и XGBoost – ассиметричные. 

В LightGBM алгоритме деревья растут по листьям - горизонтально (leaf-wise growth), 
а в XGBoost деревья растут по уровням - вертикально (level-wise growth). 

В CatBoost используется жадный алгоритм разбиения наблюдений.
<br>В LightGBM используется Gradient-based One-Side Sampling, основанный на значениях градиента для наблюдений.
<br>в XGBoost разбиение работает по предварительной сортировке значений признаков. 

На вход CatBoost могут подаваться категориальные признаки.
<br>LightGBM может принимать их на вход только в числовом формате, однако можно в порядковом виде. 
<br>XGBoost не может работать с порядковыми данными, категориальные признаки должны подаваться только после кодирования.

Все три модели поддерживают пропуски данных, для пропущенных данных можно создать категориальные бинарный указатель, и заменить пропуски, но чтобы модель понимала, что пропуски были.

Например, для работы с пропущенными значениями XGBoost в каждом решающем правиле определяется ветвь, в которую будут отправлены объекты с пропущенным значением данного признака.

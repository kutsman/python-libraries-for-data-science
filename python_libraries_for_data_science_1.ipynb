### 1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

<center>$\large precision = \LARGE \frac{TP}{TP \: + \: FP} \:\:\:\:\: \large recall = \LARGE \frac{TP}{TP \: + \: FN}$<center>

Precision показывает долю объектов относящихся к положительным и являющихся таковыми, которые определил алгоритм.
А Recall показывает долю положительных объектов, которые алгоритм нашел среди всех положительных объектов.

Таким образом в зависимости от целей можно использовать либо precision либо recall.

Например:

##### 1. Мы идем в лес собирать грибы, и нам нужно собрать как можно больше маслят на двух локациях. На 1-й локации помимо маслят еще есть подосиновики, а на второй ложные ядовитые маслята.

Таким образом на 1-й локации мы можем максимизировать recall, и он позволит собрать нам больше маслят, но и доля подосиновиков будет тоже не маленькой.
А на 2-й локации мы будем максимизировать precision, так как не смотря на количество, нам важна точность, чтобы не отравиться.

##### 2. Нам нужно выбрать место под строительство дома, и есть красивые места, но есть вероятность нарваться на болотистую почву, и у нас есть две опции: 

1) большой бюджет на строительство
2) маленький бюджет на строительство

Мы будем максимизировать precision там где бюджет маленький, чтобы случайно не нарваться на болотистый участок и не потратить весь бюджет на фундамент.
И recall классификации участков, тогда когда бюджет не ограничен.

##### 3. Если у нас есть задача диагностировать серьезное заболевание у группы пациентов, при котором необходимо начать лечение на ранней стадии, то нужно максимизировать recall.

##### 4. В гористой местности производят подрыв снега на горных вершинах, для контроля лавиноопасной обстановки, если в возможной зоне риска находятся горные деревни, то при прогнозировании результата необходимо максимизировать precision. Если же деревень нет, то будем максимизировать recall.


### 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

*F-score* - метрика, объединяющая в себе точность и полноту, выдавая как результат их гармоническое среднее, а коэффициент $\beta$, позволяет балансирвоать вклад, вносимый точностью, тем самым акцентируя внимание на той метрике, максимизация которой для нас наиболее приоритетна.

Наиболее часто встречающаяся вариация данной метрики - это F1-score, когда $\beta$ = 1, а влияние точности и полноты одинаково.

<center>$\LARGE F_{\beta} = (1+ \beta^2) \cdot \frac{precision \: \cdot \: recall}{(\beta^2 \cdot precision) \: + \: recall}$<center>

1. Регулируя 0 < $\beta$ < 1, мы можем балансировать между precision и recall. Например, полностью исключить отнесения к определенному классу.

Если же $\beta$ = 0 мы получим $F_{0}$ = precision

$F_{1} = (1+1) \cdot \frac{precision \: \cdot \: recall}{(1 \cdot precision) \: + \: recall}=2 \cdot \frac{precision \: \cdot \: recall}{precision \: + \: recall}$
